```python
#Getting list of countries (in Swedish) using a csv reader

def extracting_countries(filename):
    #Open file in read mode
    with open(filename, 'r') as read_obj:
        swe_countries = []
        #Pass the file object to reader() to get the reader object
        csv_reader = reader(read_obj)
        header = next(csv_reader)
        #print('header:', header)
        #Iterate over each row in the csv using reader object
        for row in csv_reader:
            #row variable is a list that represents a row in csv
            #print(row[2])
            if row[2] not in swe_countries: #row[2] == country name
                swe_countries.append(row[2])
    
    return swe_countries
```

```python
#Since the translation list contains some whitespaces at both initial and last position in some
#of the country names, this function will clean the data.

#This function returns a nested list with all the country names translated to English
def cleaning_trans(translation):
    first_index = []
    eng_countries = []
    for item in translation:
        #print('item:', item)
        if item[0] == ' ': #if first character in the string is a whitespace
            first_index.append(item[1:]) #add the string without the first char to the new list
        else:
            first_index.append(item) #if the word doesn't start with whitespace, just add it to the list

    for item in first_index:
        if item[-1] == ' ': #if the word ends in whitespace
            eng_countries.append(item[:-1]) #add the word without the last char to the final list
        else:
            eng_countries.append(item) #if the word doesn't have a whitespace at the end, just add the word to the list
    
    return eng_countries
```

```python
#Making a (literal) dictionary with Swedish country names as keys and their English translation as values:

def swe_eng_dict(eng_countries, swe_countries):
    replace_eng = []
    replace_swe = []

    for item in eng_countries:
        if item not in swe_countries:
            replace_eng.append(item)
    for item in swe_countries:
        if item not in eng_countries:
            replace_swe.append(item)

    #Si un país en inglés no está en la lista sueca, significa que el nombre en sueco es distinto.
    #Si un país en inglés está en la lista sueca, significa que el nombre del país es igual en inglés y en sueco.
    #Ex: Bahamas
    
    #We will use zip() and dict() to make a dictionary from the two lists above (replace_swe and replace_eng)
    #since the elements follow the same order in both list. The dict will have the Swedish names as keys and the
    #English translation as the values. This way, when we call for the key, we will get the value.

    combo = zip(replace_swe, replace_eng)
    trans_dict = dict(combo)
    
    return trans_dict
```

```python
#UPDATE: ADDING COMMAS AND NEW LINES AT THE END OF EACH COUNTRY NAME BC OF EASIER FORMATTING IN NEXT STEPS

def final_data(filename, dictionary):

    filedata = []

    with open(filename, 'r') as read_obj:
        csv_reader = reader(read_obj)
        header = next(csv_reader)
        #print('header:', header)
        filedata.append(header) #adding header = [',', 'Cities', 'Countries', 'Population']

        for row in csv_reader:
            #print('row:', row)
            country = row[2] #row[2] == country name
            if country not in dictionary:#if country name isn't in the translation dictionary
                row[0] += ','
                row[1] += ','
                row[2] += ','
                row[3] += '\n'
                filedata.append(row) #add the line [line counter, city name, country name] to the new list
            else: #if country in dictionary:
                row[0] += ','
                row[1] += ','
                row[3] += '\n'
                trans = dictionary[country] + ','
                x = [row[0], row[1], trans, row[3]] #x == [city name, english country name from dict(values)]
                #x = [row[0], trans_dict[country]] #x == [city name, english country name from dict(values)]
                filedata.append(x)
        
        filedata[0][0] = ',' #formatting the header
        filedata[0][1] = 'Cities,'
        filedata[0][2] = 'Countries,'
        filedata[0][3] = 'Population\n'

    return filedata
```

```python
def rewriting_file(filename, data):
    #data = output from final_data
    #the output is a nested list and we will convert it to string
    new_data = "".join([''.join([str(c) for c in lst]) for lst in data])
    
    with open(filename, 'w') as file:
        file.write(new_data)
```
